```python
# Create the 2x2 contingency table
# a: A=1, B=1
# b: A=1, B=0
# c: A=0, B=1
# d: A=0, B=0
a = np.sum((A == 1) & (B == 1))
b = np.sum((A == 1) & (B == 0))
c = np.sum((A == 0) & (B == 1))
d = np.sum((A == 0) & (B == 0))
```

```python
contingency_table = np.array([[a, b], [c, d]])

# Output the contingency table - look at disagreements between the models (b and c)
print("Contingency Table:")
print(contingency_table)
```

```python
# Perform McNemar's test
# asymptotic version with continuity correction
result = mcnemar(contingency_table, exact=False, correction=True)
#exact McNemar version, where b + c < 25
#result = mcnemar(contingency_table, exact=True)

# Print the result
print(f'Chi-square Statistic: {result.statistic}')
print(f'p-value: {result.pvalue}')
```
